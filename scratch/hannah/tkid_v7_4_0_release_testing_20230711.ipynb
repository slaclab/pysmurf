```python
dd={}
for channel in S.which_on(0):
    dd[channel]={}
    f, dF, sync = S.take_debug_data(band=0, channel=channel, nsamp=2**20, IQstream=0, single_channel_readout=2)
    dd[channel]['f']=f
    dd[channel]['dF']=dF
    dd[channel]['sync']=sync
    
    t = np.arange(len(f))
    t = t*1000/2.4e6
    dd[channel]['t']=t

    plt.figure()
    plt.subplot(211)
    plt.plot(t, dF)
    plt.xlabel('Time (uS)')
    plt.ylabel('Frequency error')
    plt.title(f'Channel {channel} fres={S.channel_to_freq(0,channel):.3f}')

    plt.subplot(212)
    plt.xlabel('Time (uS)')
    plt.ylabel('Tracked frequency')
    plt.plot(t, f)

    plt.figure()

    freq, pxx = signal.welch(f, nperseg=len(f)/8, fs=2.4e6)
    dd[channel]['freq']=freq
    dd[channel]['pxx_db']=10*np.log10(pxx)

    plt.semilogx(freq, 10*np.log10(pxx))
    plt.xlabel('Frequency (Hz)')
    plt.title(f'Channel {channel} fres={S.channel_to_freq(0,channel):.3f}')

plt.show()
```

Firmware will accumulate 2.4 MHz samples at 10 kHz and then stream that.  Must tell firmware what the accumulation factor is ( = 2.4 MHz / 10 kHz = 240).

```python
epics_base0 = 'smurf_server_s6:AMCc:FpgaTopLevel:AppTop:AppCore:SysgenCryo:Base[0]:'

coef       = 1/240.
coef_fixed = np.round(coef*2**15)

epics.caput(epics_base0 + 'filtSel', 0)
epics.caput(epics_base0 + 'boxcarScale', coef_fixed)
print(epics.caget(epics_base0 + 'filtSel'))
```

We had to round our coefficient for Fix16_15 data type. This will introduce a gain in streaming data of

```python
filter_gain=coef_fixed*240*2**-15
print(f'filter_gain={filter_gain}')
```

Keep this to apply to data later, although it's a very small effect.  Stream data for same length of time as the debug data.  Also return the header, which should now have the PPS pulse in it.

The PPS counter should be `counter_0`, and is 32-bit.  Right now the system is receiving a PPS pulse from a rubidium source locked to the system.  Do we see it?

```python
counter_reset_idxs=np.argwhere(np.diff(header['counter_0'])<0)
plt.plot((timestamp_s-timestamp_s[0])/1.e9,header['counter_0'])
plt.scatter((timestamp_s-timestamp_s[0])[counter_reset_idxs]/1.e9,header['counter_0'][counter_reset_idxs])

plt.ylabel('32-bit PPS counter')
plt.xlabel('Time (sec)')
```

```python
np.argwhere(np.diff(header['counter_0'])<0)
```

```python
print(f"PPS counter median = {np.median(np.diff(header['counter_0']))}")
print(f"PPS counter max = {np.max(header['counter_0'])}")
print(f"PPS counter reset dts = {np.diff(np.ndarray.flatten((timestamp_s-timestamp_s[0])[counter_reset_idxs]/1.e9))}")
```

Does this make sense?  We think the counter is incrementing at 614.4 kHz.  We're streaming at 10 kHz and downsampling by 20x, so the streamed data rate is 500 Hz.  614.4e3/500 = 1228.8.
