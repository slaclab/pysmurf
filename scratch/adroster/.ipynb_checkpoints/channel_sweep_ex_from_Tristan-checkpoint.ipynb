## Do a fine sweep centered on an arbitrary frequency

```python
freq = 4267.92
band = int(np.floor((freq - 4000) / 500))

# figure out which channel and offset
sb, centers = S.get_subband_centers(band)
assert (sb == np.arange(512)).all()
fi = np.argmin(np.abs(centers - freq + S.get_band_center_mhz(band)))
ch = S.get_channels_in_subband(band, fi)[0]
offset = freq - S.get_band_center_mhz(band) - centers[fi]
assert (np.abs(offset) < 1.2)  # channel width +- 1.2MHz

print(freq, ch, offset)
```

```python
# frequencies to sweep over
sweep_width, df_sweep = 0.3, .002
f_sweep = np.arange(-sweep_width, sweep_width, df_sweep)
n_step  = len(f_sweep)
```

### Manually sample
update the center frequency for this channel and read back the response at each fine sample

```python
%%time
# prepare the channel parameters
S.set_feedback_enable_channel(band, ch, 0)
S.set_amplitude_scale_channel(band, ch, 12)
S.set_eta_mag_scaled_channel(band, ch, 1)  # this is a bit of a hack, check

resp = np.zeros(n_step, dtype=np.complex64)
for i, df in enumerate(f_sweep):
    # set the center frequency
    S.set_center_frequency_mhz_channel(band, ch, df + offset)
    # read back real and imaginary parts
    # eta rotates between I and Q
    S.set_eta_phase_degree_channel(band, ch, 0)
    resp[i] = S.get_frequency_error_mhz(band, ch)
    # eta rotates between I and Q
    S.set_eta_phase_degree_channel(band, ch, 90)
    resp[i] += 1j * S.get_frequency_error_mhz(band, ch)
```

```python
plt.plot(f_sweep + offset, np.abs(resp))
```

### Use SerialFindFreq
Only support sweeping over all channels -- not very efficient, but may be faster if you want to measure many channels

```python
%%time
# setup sweep samples for each resonance
n_channels = S.get_number_channels(band)
resp = np.zeros(n_step, dtype=complex)
freq = np.zeros((n_channels, n_step))

# set the sweep frequencies for each channel
sweep_all[:] = f_sweep

# for the frequency of interest, set the correct offset
sweep_all[ch] += offset

# set up FindFreq
S.set_eta_scan_freq(band, sweep_all.flatten())
S.set_eta_scan_amplitude(band, 12)
S.set_run_serial_find_freq(band, 1)
```

```python
# get results
resp[:] = S.get_eta_scan_results_real(band, count=n_step*n_channels).reshape(n_channels, n_step)[ch]
resp += 1j * S.get_eta_scan_results_imag(band, count=n_step*n_channels).reshape(n_channels, n_step)[ch]
```

```python
plt.plot(f_sweep + offset, np.abs(resp))
```
